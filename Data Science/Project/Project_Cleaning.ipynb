```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\juan\\Videos\\Machine Learning\\Datasets\\Financials.csv")
df.head()
```

```python
new_columns = []
for i in range(0,16):
    new_columns.append(df.columns[i].strip())
```

```python
df.info()
```

```python
df.columns = new_columns
df.columns
```

```python
# Fungsi ini tidak bekerja dengan baik dan juga belum ditemukan kenapa dia tidak bekerja

def remove_sign(x):
    if isinstance(x, str) and x.startswith('$'):
        return x.replace('$','')
    try:
        return str(x)
    except:
        return None 
```

```python
for i in df.columns:
    df[i] = df[i].apply(remove_sign)
```

```python
df.head()
```

```python
# Kita menyelesaikan tantangan kedua setelah mencari cara menyelesaikannya dengan cara menggunakan Chat-GPT(AI)

df['Units Sold'] = df['Units Sold'].str.replace('$', '')
df.head()
```

```python
for i in df.columns:
    df[i] = df[i].str.replace('$', '')
```

```python
df.Discounts = df.Discounts.str.replace('-','')
df.head()
```

```python
for i in df.columns[4:12]:
    df[i] = df[i].astype(str)
    df[i] = df[i].str.strip()
    df[i]= pd.to_numeric(df['Units Sold'].replace(['', ' '], pd.NA).replace(',', ''), errors='coerce')
    df[i]=df[i].astype('int')
```
